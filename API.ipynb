```python
import re
def remove_emoji(data):
  if not data:
   return data
  if not isinstance(data, str):
   return data
  try:
   patt = re.compile(u'([\U00002600-\U000027BF])|([\U0001f300-\U0001f64F])|([\U0001f680-\U0001f6FF])|(\U0001F947)|(\U00002b50)|(\U0001F914)')
  except re.error:
   patt = re.compile(u'([\u2600-\u27BF])|([\uD83C][\uDF00-\uDFFF])|([\uD83D][\uDC00-\uDE4F])|([\uD83D][\uDE80-\uDEFF])')
  return patt.sub('', data)
```
